```python

# ctreating a DataFrame of Tweets.
#df=pd.DataFrame([tweet.text for tweet in nepot_tweets] , columns=['Tweets'])
#df.shape
#df

#for user_time line
df=pd.DataFrame([tweet.full_text for tweet in nepot_tweets] , columns=['Tweets'])
df.shape
df
```

```python
#Cleaning the tweets
#In the text of a tweet, there may be some unnecessary symbols which is not essential for our analysis
import re
for i in range(len(df)):
    txt = df.loc[i]["Tweets"]
    txt=re.sub(r'@[A-Z0-9a-z_:]+','',txt)#replace username-tags
    txt=re.sub(r'^[RT]+','',txt)#replace RT-tags
    txt = re.sub('https?://[A-Za-z0-9./]+','',txt)#replace URLs
    txt=re.sub("[^a-zA-Z]", " ",txt)#replace hashtags
    df.at[i,"Tweets"]=txt
    print((df['Tweets']))
df
```

```python
#creating a function to get subjectivity 
def get_sub(Tweets):
   return TextBlob(Tweets).sentiment.subjectivity
# creating function for getiing polarity
def get_pol(Tweets):
   return TextBlob(Tweets).sentiment.polarity
df['subjectivity']=df['Tweets'].apply(get_sub)
df['polarity']=df['Tweets'].apply(get_pol)
df
```

```python
# function to compute negative neutral and positive analysis

def Result(score):
    if score < 0:
        return 'Negative'
    elif score ==0:
        return 'Neutral'
    else:
        return'Positive'
df['Analysis']=df['polarity'].apply(Result)
df['Analysis']
df
```

```python
#print all the positives tweets
neu_df = df[df['Analysis']=="Neutral"]
pos_df = df[df['Analysis']=="Positive"]
neg_df = df[df['Analysis']=="Negative"]
per_neutral = (len(neu_df)/len(df))*100
per_positive = (len(pos_df)/len(df))*100
per_negative = (len(neg_df)/len(df))*100
print (per_negative , per_positive ,per_neutral)

#plotting graph
df['Analysis'].value_counts()

#plot and visualizing
plt.title('Sentimental Analysis')
plt.xlabel('Sentiment')
plt.ylabel('counts')
df["Analysis"].value_counts().plot(kind='bar')
plt.show()
```
